##### 만들어보자
요구조건
+ start, end, multiple 3개 매개변수를 이용해 start부터 시작해 multiple씩 증가해 end에 도달하면 종료하는 함수
+ 예) start = 1, end =10, multiple = 3 (출력) => 1, 4, 7

```python
def increase(start = 1, end = 10, multiple = 3):
    for i in range(start, end+1, multiple):
        if i+multiple > end or i == end:
            print(i)
        elif i+multiple < end:
            print(i, end=", ")


increase()
increase(1, 10, 2)
increase(start=0, end=50, multiple=5)
```

### Chapter 27 (재귀 함수)

```python
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

num = int(input("팩토리얼을 구하고자하는 수를 입력해주세요"))
print(f"{num}의 팩토리얼은 {factorial(num)} 입니다.")
```

```python
# 팩토리얼을 재귀함수로 구현

def oz_factorial(n):
    if n == 0:
        return 1
    else:
        return n * oz_factorial(n-1)
    
num = int(input("팩토리얼을 구하고자하는 수를 입력해주세요"))
print(f"{num}의 팩토리얼은 {oz_factorial(num)} 입니다.")
```

##### 재귀함수의 문제점  
  
피보나치 수열 : 첫째 및 둘째 항이 1이며 그 뒤에 모든 항은 바로 앞 두 항의 합인 수열로 이루어져 있습니다.
  
규칙  
+ 시작은 한쌍의 토끼
+ 두 달 이상된 토끼는 번식을 할 수 있고, 번식이 가능한 달부터 매달 새끼를 한 쌍씩 낳을 수 있음
<img src="fibo.png">

```python
# 피보나치 수열 함수 실행 시 계산되는 횟수를 파악하기 위한
# 파이썬은 함수 내부에서 함수 밖에 있는 변수를 사용할 수 없습니다.
# 그래서 global 키워드 함수를 이용해 함수 내부에서 함수 밖에 있는 변수를 사용할 수 있도록 만들어 줍니다.

count = 0

def oz_fibo(n):
    print(f'피보나치 수열{n}을 구하는 중입니다.')
    global count
    count += 1

    if n == 1:
        return 1
    if n == 2:
        return 1
    else:
        return oz_fibo(n-1) + oz_fibo(n-2)
    
num = int(input("구하고자하는 피보나치 수열의 수를 입력하시요"))
oz_fibo(num)
print(f"피보나치 수열 {num}을 구하기위해 계산된 횟수는 {count} 입니다.")
```

메모리제이션
+ 컴퓨터 프로그램이 동일한 계산을 반복적으로 해야 할 때, 이전에 계산한 값을 메모리에 저장하여 중복적인 계산을 제거호고 전체적인 실행 속도를 빠르게 해주는 기법으로 동적 계획법(DP)의 핵심이 되는 기술입니다.

코드 핵심 설명
+ 초기 값으로 설정했던 첫번째와 두번째 결과값을 memo 변수에 dict형 자료형을 넣어 저장합니다.
+ oz_fibo() 함수 실행 시 memo 변수에 계산할 피보나치 수열의 결과값이 있는지를 확인합니다.  있다면 계산을 하지 않고 이미 계산되어 저장된 결과값을 이용합니다.
+ 만일 결과값이 저장되어 있지 않은 경우 계산을 진행합니다. 그리고 계산된 결과값은 다음 계산에서 활용할 수 있도록 memo에 저장합니다.

```python
memo = {
    1 : 1,
    2 : 1
}

count = 0

def oz_fibo(n):
    print(f'피보나치 수열{n}을 구하는 중입니다.')
    global count
    count += 1

    if n in memo:
        return memo[n]
    else:
        output = oz_fibo(n-1) + oz_fibo(n-2)
        memo[n] = output
        return output
    
num = int(input("구하고자하는 피보나치 수열의 수를 입력하시요"))
oz_fibo(num)
print(f"피보나치 수열 {num}을 구하기위해 계산된 횟수는 {count} 입니다.")
```
