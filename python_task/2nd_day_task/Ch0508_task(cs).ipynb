[문제1]

Python의 변수 할당 개념을 다른 언어와 비교해 설명해주세요


Python은 C언어와 변수 할당의 개념이 다르다.  
C언어는 변수가 선언되면 메모리 주소를 할당받고 할당받은 주소로 변수의 값을 바꾸면 값이 바뀌고 주소값도 바뀌지만,  
Python은 메모리 주소로 가면 값이 아닌 객체에 대한 레퍼런스를 가지고 있어서 값을 바꾸면 주소값이 아닌 위치값이 바뀝니다.  
예를 들어 'x = 3'이라는 코드를 작성하여 변수 'x'를 선언하면 변수 'x'는 정수 '3'을 참조하게 되고, 'y = x'라고 변수 'y'를 선언하면 변수 'y'는 변수 'x'와 같이 정수 '3'을 참조하게됩니다. 그 이후 변수 'x'의 값을 변경해도 변수 'y'의 값은 변하지 않는데, 이는 변수 'x'와는 별개로 정수 '3'의 값을 참조하고 있기 때문입니다.

[문제3]

슬라이싱에 대해 간단히 설명해주세요

슬라이싱은 시퀀스 타입의 자료형(리스트, 튜플, 문자열)에서 원하는 부분만 잘나내어 이용하고 싶을 때 사용합니다.  
  
사용 방식은 시퀀스형 자료형 'x'를 잘라내어 사용하고 싶을 때,  
+ 'x[시작:끝:단계]' 형태로 사용
+ '시작' : 0 부터 시작하는 번호로 슬라이싱을 시작할 위치를 지정
+ '끝' : 1 부터 시작하는 번호로 슬라이싱을 끝낼 위치를 지정
+ '단계' : 몇 개 씩 끊어서 가져올지와 슬라이싱할 방향을 지정  

+ '시작' 또는 '끝'에 음수를 사용하면 객체들의 제일 뒤에서부터 -1을 시작으로 객체들의 앞쪽으로 이동할 수록 그 수가 점점 더 감소합니다.(-2, -3, ...)
