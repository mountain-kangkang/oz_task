[과제 4: 주문 처리 메서드 process_order]

```python
class BungeoppangShop:
    def __init__(self):
        self.stock = {
            "팥" : 10,
            "슈" : 8,
            "초" : 5
        }
        self.price = {
            "팥" : 1000,
            "슈" : 1200,
            "초" : 1500
        }
        self.sales = {
            "팥" : 0,
            "슈" : 0,
            "초" : 0
        }

    def check_stock(self):
        print("재고 확인")
        for key in self.stock.keys():
            print(f"{key}붕 : {self.stock[key]}개")
    
    def process_order(self, bread_type, bread_count):           # 단일 붕어빵만 계산 가능
        if bread_type not in self.stock.keys():
            print(f"붕어빵 맛을 잘 못 입력하셨습니다. 입력값 : {bread_type}")
        elif bread_count > self.stock[bread_type]:
            print(f"{bread_type}붕의 재고가 소진되어 판매할 수 없습니다.")
        elif bread_count <= self.stock[bread_type]:
            self.stock[bread_type] -= bread_count
            self.sales[bread_type] += bread_count
            print(f"{bread_type}붕 {bread_count}개 주문하셨습니다.")
        else:
            print("주문 확인 중 이유를 알 수 없는 오류")

    def admin_mode(self):
        pass
    def calculate_total_sales(self):
        pass

황금잉어빵 = BungeoppangShop()

bread_type, bread_count = input("구매할 붕어빵의 종류(팥, 슈, 초)와 개수를 입력해주세요. 입력예시 : 팥 1").split(" ")
황금잉어빵.process_order(bread_type, int(bread_count))
```

[과제 8: 메인 함수 main]
