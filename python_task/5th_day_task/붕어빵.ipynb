
[과제 6: 매출 계산 메서드 calculate_total_sales]

목표: calculate_total_sales 메서드를 구현하여 총 매출을 계산하고 출력합니다.

설명: self.sales와 self.prices를 사용하여 총 매출을 계산하고 출력하세요.

```python
class BungeoppangShop:
    def __init__(self):
        self.stock = {
            "팥" : 10,
            "슈" : 8,
            "초" : 5
        }
        self.price = {
            "팥" : 1000,
            "슈" : 1200,
            "초" : 1500
        }
        self.sales = {
            "팥" : 0,
            "슈" : 0,
            "초" : 0
        }

    def check_stock(self):
        print("\n재고 확인")
        for key in self.stock.keys():
            print(f"{key}붕 : {self.stock[key]}개")
        print("\n")
    
    def process_order(self):                            # 복수의 붕어빵 종류를 계산하기위한 함수 변경
        bread_type = []
        bread_count = []
        temp_bread = []
        temp_bread = list(input("구매할 붕어빵의 종류(팥, 슈, 초)와 개수를 입력해주세요. 입력예시 : 팥 1 슈 3 초 5").split(" "))
        for i in range(0, len(temp_bread), 2):
            bread_type.append(temp_bread[i])
            bread_count.append(int(temp_bread[i+1]))

        for i in range(len(bread_type)):
            if bread_type[i] not in self.stock.keys():
                print(f"붕어빵 맛을 잘 못 입력하셨습니다. 입력값 : {bread_type[i]}")
            elif bread_count[i] <= self.stock[bread_type[i]]:
                self.stock[bread_type[i]] -= bread_count[i]
                self.sales[bread_type[i]] += bread_count[i]
                print(f"{bread_type[i]}붕 {bread_count[i]}개 주문하셨습니다.")
            elif bread_count[i] > self.stock[bread_type[i]]:
                print(f"{bread_type[i]}붕의 재고가 소진되어 판매할 수 없습니다.")
            else:
                print("주문 확인 중 이유를 알 수 없는 오류")
        print("\n")

    def admin_mode(self):
        bread_type = []
        additional_stock = []
        temp_bread = []
        temp_bread = list(input("추가할 붕어빵의 종류(팥, 슈, 초)와 개수를 입력해주세요. 입력예시 : 팥 1 슈 3 초 5").split(" "))
        for i in range(0, len(temp_bread), 2):
            bread_type.append(temp_bread[i])
            additional_stock.append(int(temp_bread[i+1]))

        for i in range(len(bread_type)):
            if bread_type[i] not in self.stock.keys():
                print(f"붕어빵 종류 잘 못 입력하셨습니다. 입력값 : {bread_type[i]}")
            elif bread_type[i] in self.stock.keys():
                self.stock[bread_type[i]] += additional_stock[i]
                print(f"{bread_type[i]}붕 {additional_stock[i]}개 추가되었습니다.")
            else:
                print("재고 추가 중 이유를 알 수 없는 오류")
        self.check_stock()

    def calculate_total_sales(self):
        total_sum = 0
        for key, value in self.price.items():
            total_sum += value * self.sales[key]
            print(f"{key}붕 {self.sales[key]}개 판매,  {value * self.sales[key]}원")
        print(f"총 판매액 : {total_sum}원")
```

[과제 8: 메인 함수 main]

목표: main 함수를 작성하여 프로그램의 주요 로직을 실행합니다.

설명:
- BungeoppangShop 객체를 생성합니다.
- 사용자에게 모드를 선택하도록 요청합니다.
- "주문" 모드에서는 손님의 주문을 처리하고, "관리자" 모드에서는 재고를 추가할 수 있습니다.
- 프로그램이 종료될 때 총 매출을 계산하고 데이터를 저장합니다.


```python
def main():
    참잉어빵 = BungeoppangShop()

    mode_num = 1

    while mode_num:
        mode_num = int(input("이용할 모드를 선택하여 주십시오.\n\n 주문 : 1 \n 관리자 : 7 \n 프로그램 종료 : 0"))

        if mode_num == 1:
            참잉어빵.process_order()
        elif mode_num == 7:
            참잉어빵.admin_mode()
        elif mode_num == 0:
            print("붕어빵 판매 종료..")
        else:
            print("모드 선택 오류..!")
        
    
    참잉어빵.calculate_total_sales()     

if __name__ == "__main__":
    main()
```
